```python
from easyocr import Reader
import argparse
import cv2
import matplotlib.pyplot as plt
```

```python
# read image
image_path = "/content/drive/MyDrive/ML projects/Project_X1/Data/321.jpg"

img = cv2.imread(image_path)
```

```python
# instances text detector
Reader = Reader(['en'], gpu=True)
```

```python
# Detect text on Image
text_ = Reader.readtext(img)


```

```python
# draw box and text
for t in text_:
  print(t)
  bbox, text, score = t

  cv2.rectangle(img, bbox[0], bbox[2], (0, 255, 0), 5)

plt.imshow(img)
plt.show()
```
